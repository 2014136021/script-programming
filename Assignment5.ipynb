# 문제 3
다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)<br>

- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

>\>>> c() <br>
>\>>> d() <br>
>\>>> print(c) <br>

>\>>> print(d) <br>

- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>\>>> c = c + 5 <br>
>\>>> d = d - 5 <br>
>\>>> print(c) <br>
[Count (step: 1)] 17 <br>
>\>>> print(d) <br>
[Count (step: 2)] 9 <br>

- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>\>>> print(c) > 10 <br>
True <br>
>\>>> print(d) > 10 <br>
False <br>
>\>>> print(c) < 10 <br>
False <br>
>\>>> print(d) < 10 <br>
True <br>
>\>>> print(c) == 17 <br>
True <br>
>\>>> print(d) != 9 <br>
False <br>

```python
class Counter():
    def __init__(self, c, s=1):
        self.count=c
        self.step=s
    def __str__(self):
        return "[Count (step:%d)] %d" % (self.step,self.count)
    def incr(self):
        self.count=self.count+self.step
    def __call__(self):
        self.incr()
    def __add__(self, num):
        return Counter(self.count+num, self.step)
    def __sub__(self, num):
        return Counter(self.count-num, self.step)
    def __eq__(self, num):
        return self.count==num
    def __ne__(self, num):
        return self.count!=num
    def __gt__(self, num):
        return self.count>num
    def __lt__(self, num):
        return self.count<num
c = Counter(10) 
d = Counter(10, 2)
print (c)
print (d)
c.incr()
d.incr()
print (c)
print (d)
c()
d()
print (c)
print (d)
c=c+5
d=d-5
print (c)
print (d)
print (c > 10)
print (d > 10)
print (c < 10)
print (d < 10)
print (c == 17)
print (d != 9)
```

생성자의 인자들을 변수에 저장한다. incr를 통해 count의 값을 step에 따라 등가 시켜준다. 또한 매직 메소드들을 통해 요규사항의 연산들을 해결한다.

# 문제 4
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

init()은 생성자로 클래스가 생성될 때 불리는 함수로 변수를 초기화할 수 있다. 위의 코드에서는 인자로 들어온 리스트. 또한 eliminat_duplicate()를 호출한다.<br><br>
str()은 print 예약어나 str() 내장함수 호출하는 메소드이다. 위의 코드에서는 클래스를 print할 때 MySet:{리스트}형태로 출력하게 만든다.<br><br>
elimicate_duplicate()는 인스턴스 객체와 무관하게 클래스 이름 공간에 존재하는 메소드로서 클래스 이름을 이용하여 직접 호출할 수 있는 메소드이다. 위 코드에서는 생성자에서 호출되고 있다. 리스트 중에 중복된 값들을 제거하는 메소드이다.

# 문제 5
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

>\>>> u = s | t <br>
>\>>> print(u) <br>

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

>\>>> u = s & t <br>
>\>>> print(u) <br>
>MySet: {2, 3} <br>

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

>\>>> s = MySet([1, 2, 3])<br> 
>\>>> t = MySet([3, 4, 5]) <br>
>\>>> u = s - t <br>
>\>>> print(u) <br>
>MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __or__(self, a):##합집합
        t=[]
        for i in self:
            t.append(i)
        for j in a:
            t.append(j)
        return MySet(t)##겹치는 숫자는 생성자가 걸러줌
    
    def __and__(self, a): ##교집합
        t=[]
        for i in a:
            if(i in self):
                t.append(i)
        return MySet(t)
    def __sub__(self,a):##차집합
        for i in a:
            if(i in self):
                self.remove(i)
        return self
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
u=s|t
print(u)
h=s&t
print(h)
s=MySet([1,2,3])
t=MySet([3,4,5])
m=s-t
print(m)
```

합집합, 교집합, 차집합 모두 매직 메소드들을 이용해 만든다. 합집합은 두개의 리스트를 합치고 생성자에서 겹치는 숫자를 걸러주고 교집합은 한 리스트를 다른 리스트와 비교하여 있는 한 리스트에 숫자가 없다면 제거한다. 차집합은 한 리스트에서 다른 리스트의 숫자들을 remove해준다.
